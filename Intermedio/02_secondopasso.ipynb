```python
import pandas as pd

# Creazione del dataframe(studenti) = dataset(Tabella, Modello, Entity) 
studenti = {
    "nome": ["Mario", "Rudy", "Marco", "Luca", "Giovanni", "Francesco"],
    "cognome": ["Ricci", "Botosso", "Secci", "Bianchi", "Neri", "Rossi"],
    "email": ["mario.ricci@example.info", "rudy.botosso@example.it", "marco.secci@example.com", "luca.bianchi@example.ca", "giovanni.neri@example.com", "francesco.rossi@example.ue"],
    "eta": [30, 54, 54, 28, 35, 42],
    "citta": ["Milano", "Biella", "San Donà di Piave", "Roma", "Napoli", "Torino"],
    "voto": [8, 9, 7, 6, 8, 9],
    "stipendio": [2000, 3000, 1500, 2500, 3500, 4000]
}

# chiama pandas e crea un dataframe a partire dal dizionario studenti
df = pd.DataFrame(studenti)
df # stampa il dataframe(tipo tabella) con i dati degli studenti
```

Filtri in pandas


```python
# select * from dipendenti where stipendio > 2500
df[df["stipendio"] >= 2500] # stampa i dati degli studenti con stipendio maggiore di 2500
```

```python
# select stipendio from dipendenti where nome = 'Giovanni'
df[df["nome"] == "Giovanni"]["stipendio"] # stampa lo stipendio dello studente con nome Giovanni
```

```python
df[df['stipendio'] <= 1500][['nome', 'cognome', 'email']] # stampa nome e cognome degli studenti con stipendio uguale a 1500
```

```python
df[df['email'].str.endswith('.ue')] # stampa i dati degli studenti con email che termina con .ue
```

```python
df[df['citta'].str.contains('an')] # stampa i dati degli studenti con città che contiene la lettera a
```
