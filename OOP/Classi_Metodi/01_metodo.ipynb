```python
class Veicolo: # modello di veicolo
    """ Costruttore della calsse Veicolo
    def __init__(self):
        pass
    """
    
    def __init__(self, marca, modello): # definizione del costruttore della classe con parametro 
        self._marca = marca             # definizione dell'atributo (parametro della classe)
        self._modello = modello
    
    def info(self): # moetodo o funzione che restituisce l'informazione completa 
        return f'{self._marca} {self._modello}'
```

### Eredietà in OOP


```python
class Auto(Veicolo):
    def __init__(self, marca, modello, porte):
        super().__init__(marca, modello)
        self._porte = porte
        
    def avvia_auto(self):
        return f'La {self._marca} {self._modello} con {self._porte} porte si è acceso 🚗💨.'
```

```python
marco  = Auto("Mercedess", "Casse A", 5)

print(marco.info())
```

```python
print(marco.avvia_auto())
```

```python
class Moto(Veicolo):
    def __init__(self, marca, modello, cilindrara, carburante, colore):
        super().__init__(marca, modello)
        self._cilindrara = cilindrara
        self._carburante = carburante
        self._colore = colore
        
    def info_moto(self):
        return f"La moto {self._marca} {self._modello} {self._cilindrara} {self._carburante} {self._colore}"
```

```python
teresa = Moto("Piaggio", "Vespa", "125cc", "Benzina", "Rosso")

print(teresa.info_moto())
```
